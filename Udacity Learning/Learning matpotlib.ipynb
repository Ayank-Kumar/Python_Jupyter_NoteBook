```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [3,6,2]

plt.plot(x,y) 
plt.show()
```

```python
from matplotlib import style
```

```python
style.use('ggplot')

x = [1,2,3]
y = [3,6,2]


x2 = [2,4,6]
y2 = [3,6,2]

plt.plot(x,y,'r',label = 'first',linewidth = 4)
plt.plot(x2,y2,'g',label = 'second',linewidth = 8)

plt.legend() 
plt.show()
```

```python
plt.grid(True,color = 'k')
```

```python
style.use('ggplot')

x = [1,2,3]
y = [3,6,2]


x2 = [2,4,6]
y2 = [3,6,2]

plt.subplot(1,2,1)
plt.plot(x,y,'r',label = 'first',linewidth = 4)
plt.title('smaller range')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(1,2,2)
plt.plot(x2,y2,'g',label = 'second',linewidth = 8)
plt.title('bigger range')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid(True,color = 'k')
plt.legend() 
plt.show()
```
